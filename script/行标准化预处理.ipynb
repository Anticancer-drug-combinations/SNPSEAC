```python
import numpy as np
import pandas as pd
from scipy.stats import zscore


data1 = pd.read_csv(r'587药物对60细胞系测量矩阵.csv')
data1
```

```python
# 将第一列按英文逗号拆分成两列
data1[['DrugA', 'DrugB']] = data1['Unnamed: 0'].str.replace(r"['()]", '').str.split(',', expand=True).apply(lambda x: x.str.strip())

# 删除原有的第一列
data1.drop(['Unnamed: 0'], axis=1, inplace=True)

# 将新的两列插入到所有列之前
data1.insert(0, 'DrugB', data1.pop('DrugB'))
data1.insert(0, 'DrugA', data1.pop('DrugA'))

data1["DrugB"][0]
```

```python
data1.iloc[:, 2:] = data1.iloc[:, 2:].apply(zscore, axis=1)
data1
```

```python
data1.to_csv("587药物对60细胞系测量矩阵行标准化后.csv")
```
