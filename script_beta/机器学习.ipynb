```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# \\/r
data_gene = pd.read_csv(r'CELL_Gene.csv').set_index(["Unnamed: 0"])

# 
data_gene
```

```python
# \\/r
dataset = pd.read_csv(r'Label_synergy4000.csv').set_index(["Unnamed: 0"])

# 
dataset
```

```python
dataset=dataset.T
```

```python
columns = dataset.columns
del dataset[columns[0]]
```

```python
dict_dataset = {}
for column in dataset.columns:
    dict_dataset[column] = {}
    temp = dataset.copy()
    del temp[column]
    dict_dataset[column]["X"] = temp.values
    dict_dataset[column]["y"] = dataset[column].values
```

```python
#
temp_mean = 0
tmp_mean=0
for elem in dict_dataset:
    X = np.array(data_gene.T)
    y = dict_dataset[elem]["y"]
    
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import cross_val_predict
    from sklearn.linear_model import LinearRegression
    model  = LinearRegression()
    scores = cross_val_score(model,X, y,cv=5) #cv3cv=55
    y_pred = cross_val_predict(model,X, y,cv=5)
    print(elem)
    from sklearn import metrics
    #print("Cross validation scores:{}".format(scores))
    #print("Mean cross validation score:{:2f}".format(scores.mean()))
    print('Mean Squared Error:', metrics.mean_squared_error(y, y_pred))
    print('PCC:', abs(np.corrcoef(y, y_pred)[0,1]))
    temp_mean += metrics.mean_squared_error(y, y_pred)
    tmp_mean += abs(np.corrcoef(y, y_pred)[0,1])
mean_mean_squared_error = temp_mean/len(dataset.columns)
mean_PCC=tmp_mean/len(dataset.columns)
print(mean_PCC)
print(mean_mean_squared_error)
```

```python
import pandas as pd
import numpy as np
data1 = pd.read_excel(r'pcbi.1006752.s002.xlsx', sheet_name="Sheet1",index_col='Drug').drop('SMILES',axis=1)
data2 = pd.read_excel(r'Label_synergy4000.xlsx')
```

```python
# drugAdrugB
temp1 = []
index_l = []
for index,row in data2.iterrows():
    ct = np.array(data1.loc[row["DrugA"],:]) + np.array(data1.loc[row["DrugB"],:])
    tp = []
    for i in ct:
        if i >= 1:
            tp.append(1)
        else:
            tp.append(0)
    temp1.append(tp)
    index_l.append((row["DrugA"],row["DrugB"]))
```

```python
data_FPT=pd.DataFrame(temp1).T
```

```python
# \\/r
dataset = pd.read_csv(r'Label_synergy4000.csv').set_index(["Unnamed: 0"])

# 
dataset
```

```python
columns = dataset.columns
del dataset[columns[0]]
```

```python
dict_dataset = {}
for column in dataset.columns:
    dict_dataset[column] = {}
    temp = dataset.copy()
    del temp[column]
    dict_dataset[column]["X"] = temp.values
    dict_dataset[column]["y"] = dataset[column].values
```

```python
#
temp_mean = 0
tmp_mean = 0
for elem in dict_dataset:
    X = np.array(data_FPT.T)
    y = dict_dataset[elem]["y"]
    
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import cross_val_predict
    from sklearn.linear_model import LinearRegression
    model  = LinearRegression()
    scores = cross_val_score(model,X, y,cv=3) #cv3cv=55
    y_pred = cross_val_predict(model,X, y,cv=3)
    print(elem)
    from sklearn import metrics
    #print("Cross validation scores:{}".format(scores))
    #print("Mean cross validation score:{:2f}".format(scores.mean()))
    print('PCC:', abs(np.corrcoef(y, y_pred)[0,1]))
    temp_mean += metrics.mean_squared_error(y, y_pred)
    tmp_mean += abs(np.corrcoef(y, y_pred)[0,1])
mean_mean_squared_error = temp_mean/len(dataset.columns)
mean_PCC=tmp_mean/len(dataset.columns)
print(mean_PCC)
print(mean_mean_squared_error)
```
