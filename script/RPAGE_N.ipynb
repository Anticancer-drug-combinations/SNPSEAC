```r
# 安装和加载需要的软件包
# install.packages("plot3D")
library(plot3D)
pdf("new_o.pdf", width = 10, height = 10)

# 读取数据
data <- read.csv("587药物对60细胞系测量矩阵.csv", header=TRUE, row.names=1)
# 创建 x, y, z 向量
x <- rep(1:nrow(data), ncol(data))
y <- rep(1:ncol(data), each=nrow(data))
z <- as.vector(as.matrix(data))

# 绘制三维散点图
scatter3D(x=x, y=y, z=z, colvar=z, bty="g", pch=16, cex=1.5, 
          theta=30, phi=10, xlab="Drugs Combination", ylab="Cells", zlab="Value")
dev.off()
```

```r
# 读取数据
data <- read.csv("pcc_cpu_test_multi_cells_line.csv", header = TRUE)
pdf("p_a_scatter.pdf", width = 10, height = 10)
# 计算最小二乘回归系数
fit <- lm(data[, 5] ~ data[, 4])
a <- round(coef(fit)[2], 4)
b <- format(coef(fit)[1], scientific = FALSE, digits = 18)


# 绘制散点图和回归线
plot(data[, 4], data[, 5], xlab = "Predicted Values", ylab = "Actual Values", main = "Predicted vs. Actual Values", col = "blue")
abline(fit, col = "red")

# 添加趋势线图例
legend("topleft", legend = paste("y=", a, "x", b), col = "red", lty = 1, bty = "n")
dev.off()
```

```r
p <- summary(fit)$coefficients[, 4][2]  # 提取截距项的p值
p
```

```r
pdf("n_heatmap.pdf", width = 10, height = 10)

# 读取csv文件，将第一列设置为行名
data <- read.csv("newdata_Tanimoto.csv", row.names = 1)

# 将列名设置为行名
colnames(data) <- rownames(data)

# 将数据框转换为数值矩阵
data_matrix <- as.matrix(data)

# 定义自定义颜色
my_colors <- colorRampPalette(c("#0000FF", "#FFFFFF", "#FF0000"))

# 绘制热力图
heatmap(data_matrix, scale = "column", margins = c(10, 10), 
        col = my_colors(100))

dev.off()
```

```r
# 读取csv文件，将第一列设置为行名
data <- read.csv("newdata_Tanimoto.csv", row.names = 1)
library(plot3D)

# 提取上三角数据
data_matrix <- as.matrix(data)
data_matrix[lower.tri(data_matrix)] <- 0

# 进行t-SNE降维
tsne_result <- Rtsne::Rtsne(data_matrix, dims = 3, perplexity = 30, max_iter = 500)

# 获取降维后的三维坐标
x <- tsne_result$Y[, 1]
y <- tsne_result$Y[, 2]
z <- tsne_result$Y[, 3]
scatter3D(x=x, y=y, z=z, colvar=z, bty="g", pch=16, cex=1.5, 
          theta=30, phi=30, xlab="Drugs Combination", ylab="Cells", zlab="Value")
```
